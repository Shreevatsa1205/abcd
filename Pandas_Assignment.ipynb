# DATA1

https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv

```python
data_1 = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

```python
data_1.head()
```

```python
data_1.tail()
```

```python
# print total records and type of variables
data_1.info()
```

```python
#Print the name of all the columns.
data_1.columns
```

```python
#How is the dataset indexed?
data_1.index
```

```python
data_1['item_price']
```

```python
#How to separate $ symbol in column item_price and change its dtype to float?
data_1['item_price'] = data_1['item_price'].str.replace('$', '').astype(float)
data_1
```

```python
data_1['item_price'].dtypes
```

```python
#Create new column to get total price
data_1['total_price'] = data_1['quantity']*data_1['item_price']
```

```python
data_1
```

```python
##Which was the most ordered item? and How many items were ordered?
c = data_1.groupby('item_price')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)

```

```python
data_1[['choice_description_1']] = data_1.choice_description.str.split(expand=True)
```

```python
# delete the duplicates in item_name and quantity
filtered = data_1.drop_duplicates(['item_name','quantity'])
# select only the products with quantity equals to 1
one_prod = filtered[filtered.quantity == 1]
# select only the item_name and item_price columns
price_per_item = one_prod[['item_name', 'item_price']]
# sort the values from the most to less expensive
price_per_item.sort_values(by = "item_price", ascending = False)
```

```python
#What was the quantity of the most expensive item ordered?
data_1.sort_values(by = "item_price", ascending = False).head(1)
```

```python
#How many times were a Veggie Salad Bowl ordered
data_1[data_1.item_name == "Veggie Salad Bowl"]
```

# DATA 3

https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv

```python
data_3 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
```

```python
data_3.head()
```

```python
#Convert the type of the column Year to datetime64
data_3.Year = pd.to_datetime(data_3.Year, format='%Y')
data_3.info()
```

```python
#Set the Year column as the index of the dataframe
crime = data_3.set_index('Year', drop = True)
crime.head()
```

```python
data_3.head()
```

```python
#Return the first 3 rows of the DataFrame
data_3.head(3)
```

```python
#Return the 4 to 6 rows of the DataFrame
data_3.iloc[4:7]
```

```python
#Select just the 'Forcible_Rape' and 'Aggravated_assault' columns from the DataFrame data_3 and print
data_3.loc[:, ['Forcible_Rape', 'Aggravated_assault']].head()
```

```python
#Select the data in rows [2, 6, 8] and in columns ['Forcible_Rape', 'Aggravated_assault']
data_3.loc[data_3.index[[2, 6, 8]], ['Forcible_Rape', 'Aggravated_assault']]
```

```python
#Select only the rows where the number of murder is greater than 24,000
data_3[data_3['Murder'] > 24000]
```

```python
#Select the rows the murder is between 20k and 24k (inclusive)
data_3[data_3['Murder'].between(20000, 24000)]
```

```python
#Calculate the mean murder for each different year in data_3
data_3.groupby('Year')['Murder'].mean()
```
